```python
import numpy as np
import pandas as pd
from tqdm import tqdm
from sklearn.preprocessing import MinMaxScaler
```

```python
one = pd.read_csv('LPOINT_BIG_COMP_01_DEMO.csv') # 고객 성별
two = pd.read_csv('LPOINT_BIG_COMP_02_PDDE.csv')
three = pd.read_csv('LPOINT_BIG_COMP_03_COP_U.csv')
four = pd.read_csv('LPOINT_BIG_COMP_04_PD_CLAC.csv')
five = pd.read_csv('LPOINT_BIG_COMP_05_BR.csv')
six = pd.read_csv('LPOINT_BIG_COMP_06_LPAY.csv')
```

```python
a=pd.merge(two,four,on='pd_c',how='left')
a1=a[['cust','clac_hlv_nm']]
a2=pd.get_dummies(a1,columns=['clac_hlv_nm']) # one-hot encoding
```

```python
count_a2=a2.groupby('cust').count()
count_a3=count_a2.rename(columns={'clac_hlv_nm_가구':'count'})
count_a4=count_a3[['count']]
```

```python
a3=a2.groupby('cust').sum()
count=pd.merge(a3,count_a4,on='cust',how='left')
index=count[count['count']==5].index # 갯수 지정 
count.drop(index,inplace=True)
count.drop(['count'],axis=1,inplace=True)
a3=count
```

```python
at=a3.transpose()
```

```python
minmax_scaler = MinMaxScaler()

# 분포 저장
minmax_scaler.fit(at)

# scaling
minmax_scaled = minmax_scaler.transform(at)
```

```python
at1=pd.DataFrame(minmax_scaled)
a4=pd.DataFrame(at1.transpose())
```

```python
a4.index=a3.index
```

```python
a4.columns=a3.columns
```

```python
corr_df = a4.corr()
corr_df = corr_df.apply(lambda x: round(x ,2)) 
corr_df
```
